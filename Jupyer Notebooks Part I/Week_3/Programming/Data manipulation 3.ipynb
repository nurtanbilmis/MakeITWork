# Pandas: slicing & indices
- Voer onderstaande cel uit om het dataframe stf aan te maken.

```python
import pandas as pd
features = pd.read_csv('features.csv', parse_dates = ['Date'], dayfirst = True)
store = pd.read_csv('stores.csv',)
stf = pd.merge(store,features, how='left', left_on='Store', right_on = 'Store')
```

## 1 Inspecteren
- Inspecteer het dataframe met de 3 basisfuncties voor inspecteren.
    - Waar gaat deze dataset over?
    - Welke datatypes zijn er in het dataframe?
    - Wat is de laagste temperatuur in het dataframe?
    - Hoeveel regels heeft het dataframe?

```python
print(features.head())
print(features.info())
print(features.shape)
print("_____________________________________}")
print()
print(store.head())
print(store.info())
print(store.shape)
print("_____________________________________}")
print()
print(stf.head())
print(stf.info())
print(stf.shape)
print("_____________________________________}")
```

```python
# Welke datatypes zijn er in het dataframe?
print(f"These are the datatypes in the dataframe= \n{stf.dtypes}")
# >>> int, float, bool, datetime, object

# Wat is de laagste temperatuur in het dataframe?
print(f"\nLowest temperature in the dataframe is = {stf.Temperature.min()}\n") 

# Hoeveel regels heeft het dataframe?
print(f"Dataframe has {stf.shape[0]} rows and {stf.shape[1]} columns.") 
```

## 2 Instellen van de index en verwijderen van de index.
- Toon een nieuw dataframe stf_type met "Type" als index.

```python
stf_type = stf.set_index("Type")
display(stf_type)
```

- Verwijder de index zonder de inhoud te verwijderen.

```python
stf_type.reset_index(inplace=True)
display(stf_type)
```

## 3 Filteren
- Maak de list st1_5 met daarin de getallen (integers) 1 tot en met 5.
- Gebruik deze list om een nieuw dataframe stf_1_5 te tonen met daarin alleen de rijen van de winkels 1 tot en met 5.

```python
stf_1_5 = stf[stf.Store <= 5]
display(stf_1_5)

st1_5 = [1, 2, 3, 4, 5]
stf_1_5 = stf[stf.Store.isin(st1_5)]
display(stf_1_5)
```

## 4 Filteren op index
- Maak het dataframe stf_ind_st met als index 'Store'.
- Toon het dataframe stf_1_5_ind met daarin de rijen van winkels 1 tot en met 5.

```python
stf_ind_st = stf.set_index("Store")
stf_1_5_ind = stf_ind_st.loc[1:5]
display(stf_1_5_ind)
```

## 5 Filteren op hierarchische index
- Maak stf_ind_date_store met 'Date' en 'Store' als index. Sorteer de index oplopend.
- Toon stf_2010_02_12_1_5 met daarin de rijen van 12 februari 2010 van de winkels 1 tot en met 5.

```python
stf_ind_date_store = stf.set_index(["Date", "Store"]).sort_index()
display(stf_ind_date_store)

stf_2010_02_12_1_5 = stf_ind_date_store.loc[("20100212", 1):("20100212",5)]
display(stf_2010_02_12_1_5)
```

## 6 Filteren op delen van een datum
- Maak stf_ind_date vanuit stf met 'Date' als index. Sorteer de index oplopend.
- Toon stf_2012_1hy met daarin alleen regels uit het eerste half jaar van 2012.

```python
stf_ind_date = stf.set_index("Date").sort_index()
display(stf_ind_date)

stf_2012_1hy = stf_ind_date.loc["20120101":"20120331"]
display(stf_2012_1hy)
```

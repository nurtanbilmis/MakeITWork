```python
df2000 = df[(df['Perioden'] == 2000)
            & (df['Type onderwijs'] == 'Tertiair onderwijs|Wetenschappelijk onderwijs')
            & (df['Type uitgave'] == 'Overheidsuitgaven aan onderwijs')]

df2010 = df[(df['Perioden'] == 2010)
            & (df['Type onderwijs'] == 'Tertiair onderwijs|Wetenschappelijk onderwijs')
            & (df['Type uitgave'] == 'Overheidsuitgaven aan onderwijs')]

verschil = df2010['Waarde'].values[0] - df2000['Waarde'].values[0]

verschil

```

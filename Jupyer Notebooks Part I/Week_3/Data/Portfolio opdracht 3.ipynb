```python
import pandas as pd

df = pd.read_csv("Dataset1 week 9.csv", sep=";")
df
```

```python
# Hoeveel banen waren er in 2018? 

banen_2018 = df[(df.Perioden == 2018) & (df.Onderwerp == "Banen")]["Waarde"].sum()
print(f"Er waren {banen_2018} aantal banen in 2018")
```

```python
# Wat is in de jaren 2013-2018 het maximaal aantal personen werkzaam geweest in de Landbouw, bosbouw en visserij? 
# df.head()
# lbf_13_18 = df[(df["Bedrijfstakken/branches (SBI 2008)"] == "A Landbouw, bosbouw en visserij")
#                   & (df["Onderwerp"]=="Werkzame personen")]["Waarde"].max()

# print(f"{lbf_13_18 * 1000} aantal personen")


#GROUPBY!
filtered = df[(df["Bedrijfstakken/branches (SBI 2008)"]=="A Landbouw, bosbouw en visserij") &
              (df["Onderwerp"]=="Werkzame personen")]
grouped = filtered.groupby(["Perioden"])["Waarde"].sum().max()

print(f"\n{grouped* 1000} maximaal aantal personen zijn werkzaam geweest in de L.B.V.")
```

```python
# Werken er meer mannen of meer vrouwen in de gezondheidszorg?
display(df['Bedrijfstakken/branches (SBI 2008)'].unique())

zorg = df[(df['Bedrijfstakken/branches (SBI 2008)'] == 'Q Gezondheids- en welzijnszorg')
         &(df["Onderwerp"] == "Werkzame personen")]

test = zorg.groupby(["Geslacht"])["Waarde"].sum().reset_index()
# Er werken 26971 vrouwen en 6942 mannen, dus meer vrouwen
display(test)

if test.iloc[1,1] > test.iloc[0,1]:
    print("Meer vrouwen")
elif test.iloc[1,1] < test.iloc[0,1]:
    print("Meer mannen")
else:
    print("Gelijk aantallen")
```

```python
# Wat is de procentuele stijging in arbeidsjaren tussen 2015 en 2014? 
# df.head()
test = df[((df.Perioden == 2014) | (df.Perioden == 2015)) 
         & (df.Onderwerp == "Arbeidsjaren")]

calculate = test.groupby(["Perioden"])["Waarde"].sum().reset_index()
result = calculate.iloc[1,1] / calculate.iloc[0,1]
print(f"Procentuele stijging in arbeidsjaren tussen 2015 en 2014 is {result * 100 -100:.2f}%\n")

print(result)
```

```python
""" 
lastdf = df[(df.Perioden == 2013) | (df.Perioden == 2018)]
to_calculate = lastdf.groupby(["Perioden", "Bedrijfstakken/branches (SBI 2008)"])["Waarde"].sum().reset_index()
print(to_calculate)
print(f"\nWerknemer aantallen zijn {to_calculate.iloc[2,2]/to_calculate.iloc[0,2]:.2f}% gestegen tussen 2013 en 2018")
print(f"\nZelfstandigen aantallen zijn {to_calculate.iloc[3,2]/to_calculate.iloc[1,2]:.2f}% gestegen tussen 2013 en 2018")
"""

#Welk type banen (werkenden en zelfstandigen) is het meest gestegen in de periode 2013-2018? 

data_2013 = df[(df["Perioden"] == 2013) & (df["Onderwerp"] == "Banen")]
data_2018 = df[(df["Perioden"] == 2018) & (df["Onderwerp"] == "Banen")]

#merging both frames to work on one frame
total = data_2013.merge(data_2018, on= "Bedrijfstakken/branches (SBI 2008)", how="inner", suffixes= ("_13", "_18"))

#taking just the relevant columns to calculate
result = total[["Bedrijfstakken/branches (SBI 2008)", "Waarde_13", "Waarde_18" ]]

# getting the sum of columns per branche in the available columns Waarde_13 and Waarde_18
calculate = result.groupby(["Bedrijfstakken/branches (SBI 2008)"]).sum()

display(calculate)
#calculating difference and adding Verschil column 
calculate["Verschil"] = calculate["Waarde_18"] - calculate["Waarde_13"]

# assigning the answer to question to eind_result
eind_result = calculate[calculate["Verschil"] == calculate["Verschil"].max()]

print(f" Type {eind_result.index[0]}, zijn meest gestegen in de periode 2013-2018")
```

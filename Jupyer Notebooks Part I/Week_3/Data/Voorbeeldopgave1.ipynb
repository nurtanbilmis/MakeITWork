# Excels

Excels werken eigenlijk net zo makkelijk als csv's.

```python
excel_data1 = pd.read_excel('Dataset3.xlsx')
excel_data1
```

Het wordt iets lastiger als de Excel bijvoorbeeld geen koppen heeft of een aantal lege regels voordat de tabel begint. Dit kan overigens ook voor csv bestanden gelden.

```python
excel_data2 = pd.read_excel('Dataset4.xlsx')
excel_data2
```

De eerste regel wordt nu gezien als de kolomnaam. Check maar eens excel_data2.columns. Gelukkig valt ook dit makkelijk op te lossen door aan te geven dat er geen kolomnamen in staan met header=None

```python
excel_data3 = pd.read_excel('Dataset4.xlsx', header=None)
excel_data3
```

Zoals je ziet worden de kolommen nu genummerd. Probeer nu de code hieronder eens.

```python
excel_data4 = pd.read_excel('Dataset5.xlsx')
excel_data4
```

Dit gaat duidelijk mis omdat de excel nog lege regels bevat. Gelukkig kan ook dit makkelijk opgelost worden

```python
excel_data5 = pd.read_excel('Dataset5.xlsx', skiprows=3)
excel_data5
```

Het kan ook zo zijn dat een tabel op een ander tabblad staat in excel in dat geval kan je dit aanpassen met het agrument sheet_name

```python
excel_data6 = pd.read_excel('Dataset6.xlsx', sheet_name = 'Blad1', skiprows=3)
excel_data6
```

## API's

In de les werden ook API's behandeld. We gaan nu een kort voorbeeld zien van het automatisch ophalen van data vanuit het CBS. Zorg dat de library cbsodata geinstalleerd is op je computer

```python
# Importeer library
import cbsodata
# Met dit commando haal je alle tabellen op van het CBS, deze slaan we nu op in een lijst
tables = cbsodata.get_table_list()
# Print de informatie van de eerste tabel
print(tables[0])
```

Zoals je ziet is de informatie van de tabel opgeslagen op een ingewikkelde manier in een dictionary. We gaan nu niet in op hoe dat precies gebruikt wordt maar we gaan de data uit de tabel ophalen. Dit kan even duren. Vergelijk het resultaat met de tabel zelf op https://opendata.cbs.nl/statline/#/CBS/nl/dataset/84669NED/table?ts=1579872202002

```python
api_data = cbsodata.get_data('84669NED')
api_data
```

Het is een beetje lastig om te lezen maar de data wordt nu als een lijst teruggegeven waarbij ieder element in de lijst weer een verzameling (dictionary) is met de kolomnaam van de tabel met een waarde. Gelukkig kunnen we lijst makkelijk omzetten naar een georganiseerde tabel

```python
api_data_tabel = pd.DataFrame(api_data)
api_data_tabel
```

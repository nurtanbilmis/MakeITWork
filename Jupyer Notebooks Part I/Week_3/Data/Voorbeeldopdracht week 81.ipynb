## Requests

Het wordt iets ingewikkelder wanneer je geen package hebt maar wanneer je met requests moet werken. In dat geval heb je direct met de API

```python
import requests
response = requests.get("https://jsonplaceholder.typicode.com/todos/1")
response
```

De response zelf zegt nog iet zo veel, Het is een response object. De 200 staat voor het feit dat een succesvolle request is geweest. Lasten we eens kijken wat we met een response kunnen doen. Hieronder staat een lijst met alle attributen van dit object

```python
response.__dict__.keys()
```

We kunnen ook de data ophalen

```python
response.json()
```

Dezelfde methode werkt natuurlijk ook voor de OData API van CBS.
In de code hieronder halen we twee tabellen op.    
De eerste is een tabel met verwijzingen naar metadata, de tweede tabel is exact dezelfde tabel als wij met de cbsodata module hebben opgehaald.  
De waardes van deze tabel zijn echter gecodeerd en kunnen worden aangepast (zodat zij begrijpbaar zijn) door de waardes uit de tabel te koppelen aan de metadata.  
Hoe dit precies werkt leer je in een toekomstige les.  

```python
#Eerste tabel: metadata
response = requests.get('https://odata4.cbs.nl/CBS/84669NED') 
df_cbs = pd.DataFrame(response.json()['value'])
display(df_cbs)

#Tweede tabel: waarnemingen
response = requests.get('https://odata4.cbs.nl/CBS/84669NED/Observations') 
df_cbs_observations = pd.DataFrame(response.json()['value'])
display(df_cbs_observations)
```

## Ongestructureerde data

Hieronder werken we met een klein voorbeeld van web scraping. We werken opnieuw met requests.
Het recept is vrij simpel, we gebruiken opnieuw get om een webpagina op te halen. Van die pagina kunnen we dan de text opvragen

```python
import requests

URL = "https://realpython.github.io/fake-jobs/"
page = requests.get(URL)

print(page.text)
```

Het lastige is dat we dan slechts een html code krijgen waar we niet heel veel mee kunnen. Gelukkig kunnen we dat ook verhelpen met een andere library: Beautifulsoup. 

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, "html.parser")
soup
```

Dit lijkt hetzelfde maar er is een groot verschil. We kunnen dit object bevragen. Zo kunnen we op html_id zoeken. 

```python
results = soup.find(id="ResultsContainer")
results
```

Een andere optie is om specifieke classes te zoeken. In dit geval card_content

```python
job_elements = results.find_all("div", class_="card-content")
job_elements
```

```python
import pandas as pd
onderwijs_uitgave = pd.read_excel("onderwijs_uitgave.xlsx", na_values=".")

onderwijs_uitgave
```

```python
# Bepaal door het selecteren van data uit een dataset in Python de uitgaven aan primair basisonderwijs in 1989.

onderwijs_1989 = onderwijs_uitgave[(onderwijs_uitgave.Perioden == 1989)&
                                  (onderwijs_uitgave["Type onderwijs"].str.startswith("(Pre)primair"))]
onderwijs_1989

```

```python
# Bepaal door middel van Python code de toename van de overheidsguitgaven aan onderwijs voor het onderwijstype 
# wetenschappelijk onderwijs in de periode 2000-2010

science_2000_2010 = onderwijs_uitgave[(onderwijs_uitgave.Perioden >=2000) & (onderwijs_uitgave.Perioden <= 2010) & 
                    (onderwijs_uitgave["Type onderwijs"] == "Tertiair onderwijs|Wetenschappelijk onderwijs") &
                    (onderwijs_uitgave["Type uitgave"] == "Overheidsuitgaven aan onderwijs")]

print(f"\nToename is = {science_2000_2010.iloc[10,3] - science_2000_2010.iloc[0,3]}")

import matplotlib.pyplot as plt
display(science_2000_2010)
science_2000_2010.plot(x="Perioden")
```

```python
# Selecteer alleen de overheidsuitgaven aan onderwijs in de kolom Type uitgaven 
# sla de data op in een tabel met de naam â€œoverheidsuitgaven
onderwijs_uitgave

overheidsuitgaven = onderwijs_uitgave[onderwijs_uitgave["Type uitgave"] == "Overheidsuitgaven aan onderwijs"]
overheidsuitgaven
```

```python
# Gebruik de nieuwe tabel en maak een draaitabel met in de rijen de Perioden in de kolommen het type onderwijs. 
# Welke manier van aggregeren wordt automatisch gekozen?
pd.pivot_table(overheidsuitgaven, index='Perioden', columns='Type onderwijs', values='Waarde')

# mean wordt automatisch gekozen
```

```python
# Maak nu een draaitabel met perioden in de rijen, geen kolommen en als aggregeerfunctie â€˜maxâ€™. 
# Dit geeft per jaar de maximale uitgave per type onderwijs.
pd.pivot_table(overheidsuitgaven, index="Perioden", aggfunc=max)
```
